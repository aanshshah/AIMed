```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, average_precision_score, precision_recall_curve, roc_curve, auc, precision_score, roc_curve, confusion_matrix
```

```python
def classifaction_report_csv(report, filename):
    report_data = []
    lines = report.split('\n')
    for line in lines[2:4]:
        if not line: continue
        row = {}
        row_data = line.split('      ')
        row['class'] = row_data[0]
        row['precision'] = float(row_data[1])
        row['recall'] = float(row_data[2])
        row['f1_score'] = float(row_data[3])
        row['support'] = float(row_data[4])
        report_data.append(row)
    dataframe = pd.DataFrame.from_dict(report_data)
    dataframe.to_csv(filename, index = False)
```

```python
hierarchy_path = "../data/hierarchy_clustered"
dirs = os.listdir(hierarchy_path) #list all the clustering types
file_mapping = {} #maps cluster name to x to y
for directory in dirs:
    dir_path = os.path.join(hierarchy_path, directory)
    if os.path.isdir(dir_path):
        files = os.listdir(dir_path)
        files.sort()
        file_map = {}
        for file in files:
            file_type = file.split("_")[0]
            cluster_num = file.split("_")[2][:-4]
            if file_type == "X":
                file_map[cluster_num] = [file]
            else:
                curr = file_map.get(cluster_num)
                curr.append(file)
                file_map[cluster_num] = curr
        file_mapping[directory] = file_map
```

```python
done = {'depth_1', 'depth_2'}
for cluster, file_map in file_mapping.items():
    if cluster in done: continue
    for cluster_key, files in file_map.items():
        filename = cluster + ' ' + cluster_key+'.csv'
        cluster_dir_path = os.path.join(hierarchy_path, cluster)
        X_path = os.path.join(cluster_dir_path, files[0])
        try:
            X = pd.read_csv(X_path)
        except:
            continue
        if cluster in done: X = X.drop(['cluster'], axis=1)
        Y_path = os.path.join(cluster_dir_path, files[1])
        try:
            Y = pd.read_csv(Y_path, header=None)
        except:
            continue
        optimal_model = XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.4, gamma=0.0,
              learning_rate=0.1, max_delta_step=0, max_depth=3,
              min_child_weight=9, missing=None, n_estimators=100, n_jobs=1,
              nthread=4, objective='binary:logistic', random_state=0,
              reg_alpha=0.25, reg_lambda=1, scale_pos_weight=8.286036036036036,
              seed=1, silent=None, subsample=0.8, verbosity=1)
        try:
            X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
        except:
            continue
        optimal_model.fit(X_train, y_train)
        y_preds = [0 if x < 0.5 else 1 for x in optimal_model.predict(X_test)]
        report = classification_report(y_test, y_preds, target_names=['Not Readmitted', 'Readmitted'], labels=[0, 1])
        classifaction_report_csv(report, filename)        
        
```

```python
#calculate mean and median 
def calculate_mean_and_median(cluster_directory_name):
    readmit_str = '    Readmitted'
    files = os.listdir(cluster_directory_name)
    files = [file for file in files if file != '.DS_Store']
    n = len(files)
    f1 = {'Not Readmitted': [], readmit_str: []}
    precision = {'Not Readmitted': [], readmit_str: []}
    recall = {'Not Readmitted': [], readmit_str: []}
    support = {'Not Readmitted': [], readmit_str: []}
    scores = ['f1_score', 'precision', 'recall', 'support']
    values = [f1, precision, recall, support]
    metric_map = dict(zip(scores, values))
    for file in files:
        file = os.path.join(cluster_directory_name, file)
        results = pd.read_csv(file, names=scores, header=None,skiprows=[0])
        for metric in scores:
            readmit = results[metric][readmit_str]
            curr = metric_map[metric][readmit_str]
            curr.append(readmit)
            metric_map[metric][readmit_str] = curr
            not_readmit = results[metric]['Not Readmitted']
            curr = metric_map[metric]['Not Readmitted']
            curr.append(not_readmit)
            metric_map[metric]['Not Readmitted'] = curr
    #write the means and medians to a csv
    mean = {}
    median = {}
    for score in scores:
        mean[score] = {'Not Readmitted': [], readmit_str: []}
        median[score] = {'Not Readmitted': [], readmit_str: []}
    for metric, values in metric_map.items():
        mean[metric]['Not Readmitted'] = sum(values['Not Readmitted'])/n
        curr = values['Not Readmitted']
        curr.sort()
        mid = len(values['Not Readmitted'])//2
        median[metric]['Not Readmitted'] = curr[mid]

        mean[metric][readmit_str] = sum(values[readmit_str])/n
        curr = values[readmit_str]
        curr.sort()
        mid = len(values[readmit_str])//2
        median[metric][readmit_str] = curr[mid]
            
    mean_df = pd.DataFrame.from_dict(mean)
    mean_df.to_csv(os.path.join(cluster_directory_name, 'mean_results.csv'))
    median_df = pd.DataFrame.from_dict(median)
    median_df.to_csv(os.path.join(cluster_directory_name, 'median_results.csv'))
```

```python
calculate_mean_and_median('results/kmodes_clusters')
```
